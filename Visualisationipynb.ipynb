Bar Plot pour compoarer les ventes globales par plateform



```python
platform_sales = data_interpolated.groupby('Platform')['Global_Sales'].sum().reset_index()
#Trier les donn√©es par ventes globales
platform_sales = platform_sales.sort_values(by='Global_Sales', ascending=True)

plt.figure(figsize=(12, 6))
sns.barplot(x='Global_Sales', y='Platform', data=platform_sales, palette='viridis')
plt.title('Ventes Globales par Plateforme', fontsize=16)
plt.xlabel('Ventes Globales (en millions)', fontsize=12)
plt.ylabel('Plateforme', fontsize=12)

plt.show()
```

On remarque que (NG, WS, TG16, 3DO, GG, PCFX) sont des platformes finie et pour cela on raison que n'est pas important de fourni des jeux pour ces platforme  

Bar Plot pour comparer les vente par genre

```python
platform_sales = data_interpolated.groupby('Genre')['Global_Sales'].sum().reset_index()
#Trier les donn√©es par ventes globales
platform_sales = platform_sales.sort_values(by='Global_Sales', ascending=True)

plt.figure(figsize=(12, 6))
sns.barplot(x='Global_Sales', y='Genre', data=platform_sales, palette='viridis')
plt.title('Ventes Globales par Genre', fontsize=16)
plt.xlabel('Ventes Globales (en millions)', fontsize=12)
plt.ylabel('Plateforme', fontsize=12)

plt.show()
```

Comparer les ventes par r√©gion

On remarque que les jeux (action, sports, shooter,role-playing, platform) sont les plus jeux rentable

```python
# Pr√©parer les donn√©es pour le graphique √† barres group√©es
sales_data = data_interpolated[['Name', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
sales_data.set_index('Name', inplace=True)

# D√©finir la largeur des barres
bar_width = 0.2
index = np.arange(len(sales_data))

# Cr√©er un graphique √† barres empil√©es
sales_data.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='viridis')

# Ajouter des titres et des labels
plt.title('Ventes par R√©gion (Empil√©es)', fontsize=16)
plt.xlabel('Jeux', fontsize=12)
plt.ylabel('Ventes (en millions)', fontsize=12)
plt.legend(title='R√©gions', bbox_to_anchor=(1.05, 1), loc='upper left')

# Afficher le graphique
plt.tight_layout()
plt.show()
```

Nuage de point pour visualiser la relation entre les ventes en NA_sales and EU_sales

```python
# Cr√©er le nuage de points
plt.figure(figsize=(10, 6))
plt.scatter(data_interpolated['NA_Sales'], data_interpolated['EU_Sales'], alpha=0.7)

# Ajouter des titres et des labels
plt.title('Relation entre les ventes en NA et en EU', fontsize=16)
plt.xlabel('Ventes en Europe (EU Sales)', fontsize=12)
plt.ylabel('Ventes en Am√©rique du Nord (NA Sales)', fontsize=12)

# Afficher la grille
plt.grid()

# Afficher le graphique
plt.show()
```

On remarque que les points sont tr√®s concentr√©s dans une zone particuli√®re du graphique c-a-d la majorit√© des ventrs se situent dans ce plage,et les pooints se regrouper le long d'une ligne ascendente c-a-d une corr√©lation positive entre les ventes en NA et EU , en fin je remarque qu'il ya une point tr√®s √©loign√©s indique des jeux qui sont un succ√©s exceptionnel dans une r√©gion et non dans l'autre

Nuage entre JP et EU

```python
# Cr√©er le nuage de points
plt.figure(figsize=(10, 6))
plt.scatter(data_interpolated['JP_Sales'], data_interpolated['EU_Sales'], alpha=0.7)

# Ajouter des titres et des labels
plt.title('Relation entre les ventes en JP et en EU', fontsize=16)
plt.xlabel('Ventes en Europe (EU Sales)', fontsize=12)
plt.ylabel('Ventes en Japon (JP Sales)', fontsize=12)

# Afficher la grille
plt.grid()

# Afficher le graphique
plt.show()
```

```python
total_sales = {
    'NA_Sales': data_interpolated['NA_Sales'].sum(),
    'EU_Sales': data_interpolated['EU_Sales'].sum(),
    'JP_Sales': data_interpolated['JP_Sales'].sum(),
    'Other_Sales': data_interpolated['Other_Sales'].sum()
}
labels = total_sales.keys()
sizes = total_sales.values()
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = (0.1, 0, 0, 0)

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('R√©partition des ventes de jeux vid√©o par r√©gion', fontsize=16)
plt.axis('equal')  # Pour s'assurer que le diagramme est un cercle
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supposons que vous avez d√©j√† charg√© et nettoy√© votre dataset
# data_interpolated = pd.read_csv('path_to_your_interpolated_dataset.csv')

# V√©rifiez les types de donn√©es et les valeurs manquantes
print(data_interpolated.dtypes)
print(data_interpolated.isnull().sum())

# Cr√©er un DataFrame long pour les ventes
sales_data = pd.melt(data_interpolated,
                     value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'],
                     var_name='Region', value_name='Sales')

# Cr√©er le box plot avec Seaborn
plt.figure(figsize=(12, 6))
sns.boxplot(x='Region', y='Sales', data=sales_data)

# Personnaliser le graphique
plt.title('Box Plot of Video Game Sales by Region')
plt.ylabel('Sales (in millions)')
plt.xlabel('Region')
plt.grid(axis='y')

# Afficher le graphique
plt.show()
```

Je remarque l'abscence de boite le Q1 , Q3 , IQR et whiskers ce qui signiifie que les donn√©e sont tr√©s homog√©nes , les points qui sont en dehors indique des jeux qui en un succ√©s exp ou des echecs notables , les boites pour les vnetes en NA est plus grand que Other_sales cela indique les jeux sont vendent en NA mieux
