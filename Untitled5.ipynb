Découverte des données

```python
#importation des library
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#collecter les noms des fichiers (dataset)
directory_path = '/content/'
files = os.listdir(directory_path)
csv_files = [file for file in files if file.endswith('.csv')]
for csv_file in csv_files:
    print(csv_file)
```

```python
path=directory_path
#créer une base de donnée vide
all_data=pd.DataFrame()
# Boucle pour lire et concaténer les fichiers .csv
for csv_file in csv_files:
    current_data = pd.read_csv(os.path.join(directory_path, csv_file))
    all_data = pd.concat([all_data, current_data], ignore_index=True)
print(all_data)
```

```python
all_data.to_csv(path+'all_data.csv',index=False)
```

```python
#types des objets
all_data.dtypes
```

```python
#afficher la df
all_data.head()
```

```python
#afficher les valeurs manquantes
all_data.isnull().sum()
```

```python
#supprimer les valeurs manquantes
all_data=all_data.dropna(how='all')
all_data.shape
```

Quel est le mois durant lequel nous avons réalisé le meilleur chifre d'affaire?

```python
def month(x):
  return x.split('/')[0]
```

```python
all_data['Month']=all_data['Order Date'].apply(month)
all_data
```

```python
all_data['Month'].unique()
```

```python
all_data=all_data[all_data['Month']!='Order Date']
all_data['Month'].unique()
```

```python
all_data.dtypes
```

```python
all_data["Month"]=all_data['Month'].astype(int)
all_data.dtypes
```

```python
all_data['Price Each']=all_data['Price Each'].astype(float)
all_data['Quantity Ordered']=all_data['Quantity Ordered'].astype(int)
all_data.dtypes
```

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data
```

```python
all_data.groupby('Month')['Sales'].sum()
```

```python
months=range(1,13)
plt.bar(months,all_data.groupby('Month')['Sales'].sum())
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Month number')
plt.show()
```

En quelle moment doit on faire une compagne publicitaire pour avoir plus de vnetes?

```python
all_data['Hour']=pd.to_datetime(all_data['Order Date']).dt.hour
all_data
```

```python
keys=[]
hours=[]
for key,hour in all_data.groupby('Hour'):
  keys.append(key)
  hours.append(len(hour))
hours
```

```python
plt.grid()
plt.plot(keys,hours)
plt.xlabel('heure de la journée')
plt.ylabel('nombre des commande')
```
